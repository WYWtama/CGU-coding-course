```python
for number1 in range(1,10):
  for number2 in range(1,10):
    number1 = int(number1)
    number2 = int(number2)
    if number1 * number2 < 10 and number2 < 10:
      print(number1, "x", number2, "=", number1*number2, end = "  ")
    elif number1*number2 >= 10 and number2 < 10:
      print(number1, "x", number2, "=", number1*number2, end = " ")
  print("")
```

```python
number1 = 1;
while number1 < 10:
    number2 = 1
    while number2 < 10:
        ans = number1 * number2
        if ans < 10:
          print ('%d * %d = %d ' %(number2, number1 ,ans), end = "  ")
        else:
          print ('%d * %d = %d ' %(number2, number1 ,ans), end = " ")
        number2 += 1
    print("")
    number1 += 1
```
